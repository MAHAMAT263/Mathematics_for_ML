```python
# Histogram
plt.figure(figsize=(9,5))
plt.hist(df['co2'], bins=70 , color="red" , alpha=0.3 , label="Co2" )
plt.legend()
plt.xlabel("X-axes")
plt.ylabel("Y-axes")
plt.title("histogram that depicts the distribution of carbon Dioxide")
plt.show()
```

```python
# highway
plt.figure(figsize=(11,5))
plt.hist(df['highway'], bins=90 , color="blue" , alpha=0.5 , label="highway" )
plt.legend()
plt.xlabel("X-axes")
plt.ylabel("Y-axes")
plt.title("histogram that depicts the distribution of highway")
plt.show()
```

```python
# comb
plt.figure(figsize=(11,6))
plt.hist(df['comb'], bins=100 , color="black" , alpha=0.6 , label="comb" )
plt.legend()
plt.xlabel("X-axes")
plt.ylabel("Y-axes")
plt.title("histogram that depicts the distribution of comb")
plt.show()
```

```python
# UHighway
plt.figure(figsize=(12,3))
plt.hist(df['UHighway'], bins=90 , color="green" , alpha=0.8 , label="UHighway" )
plt.legend()
plt.xlabel("X-axes")
plt.ylabel("Y-axes")
plt.title("histogram that depicts the distribution of UHighway")
plt.show()
```

```python
# city
plt.figure(figsize=(10,5))
plt.hist(df['city'], bins=50 , color="purple" , alpha=0.9 , label="city" )
plt.legend()
plt.xlabel("X-axes")
plt.ylabel("Y-axes")
plt.title("histogram that depicts the distribution of city")
plt.show()
```

```python
df.drop(['feScore', 'comb'], axis=1, inplace=True , errors="ignore")
df_numeric = df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(12, 8))
sns.heatmap(df_numeric.corr(), vmin=-1, vmax=1, center=0, cmap = sns.diverging_palette(20, 220, n=500), square=True)
plt.title('Correlation Heatmap')
plt.show()
plt.savefig('heatmap_plot.png')
```

Interpretation of the Correlation Heatmap:

-The 'co2' column shows a strong positive correlation with the 'comb' column (fuel efficiency), which is in line with expectations.

-The 'feScore' and 'ghgScore' columns are highly correlated with 'co2', suggesting they may not offer additional insights when predicting the GHG score.

-The 'city', 'UCity', 'highway', and 'UHighway' columns also exhibit a strong positive correlation with 'co2' and fuel efficiency scores.
